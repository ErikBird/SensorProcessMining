```python
import matplotlib.pyplot as plt
from ProcessSimulator.src.simulator.signal_generator import perlin_noise
from utils import get_assets_path, get_matplotlib_size, get_color_hex_list
import copy
from ProcessSimulator.src.simulator.signal_generator import add_trend, add_noise, add_spike, add_spike_cluster, add_increase_by_percent, add_stuck_at_zero, add_stuck_at_constant

linewidth = 5.0
save = True
```

```python
signal = perlin_noise(amplitude = 1, wavelength= 400, octaves= 5, divisor = 2, num_points = 300, fixed=20, variable=20)

fig, ax = plt.subplots(1, 1, figsize=get_matplotlib_size(418.25555))
ax.plot(signal, linewidth=linewidth)
ax.set_ylim(ymin=0, ymax=45)
fig.tight_layout()
plt.show()

if save: fig.savefig(get_assets_path()/'SyntheticDatasetGenerator'/'SensorAnomalies'/"signal.pgf", format='pgf')
```
